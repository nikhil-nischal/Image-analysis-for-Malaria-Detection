```
import csv

path="gdrive/My Drive/Codes/shivank/CV/"
f = open(path+'submission.csv', 'r')
reader = csv.reader(f)

w = open(path+'submission1.csv', 'w+')
writer = csv.writer(w)
for row in reader:
  name = row[1]
  try:
    img = open_image(root_path+name)
    pred_class,pred_idx,outputs = learn.predict(img)
    print(pred_class)
    if str(pred_class)=='Parasitized':
      row[2]=0
    else:
      row[2]=1
    print(row)
    writer.writerow(row)
  except:
    print(row)
    writer.writerow(row)

f.close()
w.close()
```
